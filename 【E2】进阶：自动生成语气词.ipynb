## 导入工具包

```python
import torchaudio
import torch
from ChatTTS import ChatTTS
import soundfile
from IPython.display import Audio
```

## 载入预训练模型

```python
chat = ChatTTS.Chat()
```

```python
# 加载默认下载的模型
chat.load(compile=False) # 设置为Flase获得更快速度，设置为True获得更佳效果
```

## 原文

```python
text = "So we found being competitive and collaborative was a huge way of staying motivated towards our goals, so one person to call when you fall off, one person who gets you back on then one person to actually do the activity with."

# text = '三星堆遗址，位于四川省广汉市，被称为二十世纪人类最伟大的考古发现之一'
```

## 自动生成语气词

```python
refine_text = chat.infer(text, refine_text_only=True)
```

```python
refine_text
```

```python
refine_text
```

> `refine_text_only=True`的作用：断句、自动添加停顿、转换阿拉伯数字等。

## 语音合成模型推理

```python
# 要生成语气词
wavs = chat.infer(refine_text)

# # 不要生成语气词
# wavs = chat.infer(text, skip_refine_text=True)
```

## 试听

```python
Audio(wavs[0], rate=24000)
```

## 导出保存为音频文件

```python
torchaudio.save("output/output_e2.wav", torch.from_numpy(wavs[0]).unsqueeze(0), 24000)
```
